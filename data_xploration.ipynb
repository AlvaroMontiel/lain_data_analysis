```python
import pandas as pd
```

```python
df = pd.read_excel('data/reporte_formularios_250212_1610.xlsx', sheet_name="reporte_formularios_250212_1610")
```

```python
df.head()
```

```python
# Seleccionar las columnas que sean de tipo 'object' o 'category'
columnas_categoricas = df.select_dtypes(include=['object', 'category']).columns

# Iterar sobre las columnas categóricas y mostrar los value_counts()
for columna in columnas_categoricas:
    print(f"Value counts para la variable '{columna}':")
    print(df[columna].value_counts())
    print("\n")  # Separador entre variables
```
